This algorithm is always picking our smallest edge out of the edges that already have a node that is connected. The reason this works for finding the minimal tree is that each iteration is finding the minimal way to add a new node visit without going to a node already visited. The way I like to think about it is imagine each edge is corresponding to the difficulty to getting to get from point $i$ to point $j$. What our algorithm is doing is trying to find the easiest difficulty to get to a new point given the points we have already gone to.

```python
adj1_matrix = [
    # a  b  c  d  e  f  g  h  i  j
    [ 0, 5, 0, 0,10,17, 0, 0, 0,28],  # a
    [ 5, 0, 8, 4, 3, 0, 0, 0, 0, 0],  # b
    [ 0, 8, 0, 9, 0, 0,19, 0, 0, 0],  # c
    [ 0, 4, 9, 0, 2, 0, 3, 0, 0, 0],  # d
    [10, 3, 0, 2, 0,13, 5, 0, 0, 0],  # e
    [17, 0, 0, 0,13, 0,14, 8, 0, 5],  # f
    [ 0, 0,19, 3, 5,14, 0,12,21, 0],  # g
    [ 0, 0, 0, 0, 0, 8,12, 0, 7, 8],  # h
    [ 0, 0, 0, 0, 0, 0,21, 7, 0,10],  # i
    [28, 0, 0, 0, 0, 5, 0, 8,10, 0],  # j
]

def find_root(parent_array, i):
  if parent_array[i] == i:
    return i
  parent_array[i] = find_root(parent_array, parent_array[i])
  return parent_array[i]

def union(parent_array, rank_array, x, y):
  root_x = find_root(parent_array, x)
  root_y = find_root(parent_array, y)
  if root_x != root_y:
        if rank_array[root_x] > rank_array[root_y]:
            parent_array[root_y] = root_x
        else:
            parent_array[root_y] = root_x
            if rank_array[root_x] == rank_array[root_y]:
                rank_array[root_x] += 1

def sort_by_weight(adj_matrix):
  weight_sorted_edges = []
  for i in range(len(adj_matrix)):
    for j in range(len(adj_matrix[i])):
      if adj_matrix[i][j] != 0:
        weight_sorted_edges.append((i, j, adj_matrix[i][j]))
  weight_sorted_edges.sort(key=lambda x: x[2])
  return weight_sorted_edges

def Kruskals(adj_matrix):
  E = sort_by_weight(adj_matrix)
  T = []
  letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

  parent = list(range(len(adj_matrix)))
  rank = [0] * len(adj_matrix)

  for u, v, weight in E:
    if find_root(parent, u) != find_root(parent, v):
      T.append((u, v, weight))
      union(parent, rank, u, v)

  T = [(letters[u], letters[v], weight) for u, v, weight in T]
  return T

print(Kruskals(adj1_matrix))
```

This algorithm goes off the idea that we are making multiple rooted trees in the process of doing Kruskal's. We update the parent of a node once we have connected them by an edge. If the parents are the same, that means the nodes are already connected by a rooted tree. The rank filters how "parent-y" the node is, so that each tree can have a decisive node to become the "Mother node". It also creates the same tree as Prim's:

```python
import numpy as np
import math
import matplotlib.pyplot as plt

def play_chomp():

  m, n = input(f'The board will be m columns and n rows, please enter the values for m and n: ').split()
  m = int(m) - 1
  n = int(n) - 1

  board = initialize_board(m, n)
  print('Board is now initialized: ')
  print_board(board)

  i = 0

  while any(0 in row for row in board):

    i += 1

    if i % 2 == 1:
      print("Player 1's turn: ")

      if n == 0: #trivial cases

        move_row = 0
        move_column = 1
        board = turn(board, move_row, move_column)
        print_board(board)

      if m == 0:
        move_row = n - 1
        move_column = 0
        board = turn(board, move_row, move_column)
        print_board(board)

      if m == n: # square board

        if i == 1: #first move

          move_row = n - 1
          move_column = 1


        else:

          move_row = n
          move_column = n - p_row

        board = turn(board, move_row, move_column)
        print_board(board)

      if n == 1: #2xn

        if i == 1:

          move_row = 0
          move_column = m
          board = turn(board, move_row, move_column)
          print_board(board)
        else:
          if p_row == 0:
            move_row = 1
            move_column = p_column + 1
            board = turn(board, move_row, move_column)
            print_board(board)

          if p_row == 1:
            move_row = 0
            move_column = p_column -1
            board = turn(board, move_row, move_column)
            print_board(board)

      if m == 1: #nx2

        if i == 1:

          move_row = 0
          move_column = 1
          board = turn(board, move_row, move_column)
          print_board(board)
        else:
          if p_column == 0:
            move_column = 1
            move_row = p_row + 1
            board = turn(board, move_row, move_column)
            print_board(board)
          if p_column == 1:
            move_column = 0
            move_row = p_row - 1
            board = turn(board, move_row, move_column)
            print_board(board)

      #else: #mxn

    else:
      invalid_input = True
      while invalid_input == True:
        print("Player 2's turn: ")
        p_row, p_column = input('Choose your row and column: ').split()
        p_row = int(p_row) - 1
        p_column = int(p_column) - 1
        if board[p_row][p_column] == 'x' or ((0 > p_row < n) or (0 > p_column < m)):
          if p_row == n and p_column == 0:
            print('You ate the poison and lost!')
            return
          else:
            print('Invalid move, please try again')
        if board[p_row][p_column] != 'x':
          invalid_input = False
          move_row = p_row
          move_column = p_column
          board = turn(board, move_row, move_column)
          print_board(board)

  print('Player 1 wins!')

def initialize_board(m, n):
  board = [[ 0 for _ in range(m + 1)] for _ in range(n + 1)]
  board[n][0] = 'x'
  return board

def print_board(board):
  for i in range(len(board)):
    for j in range(len(board[0])):
      print(board[i][j], end=' ')
    print()

def turn(board, move_row, move_column):
  n = len(board)
  m = len(board[0])

  for i in range(move_row + 1):
    for j in range(move_column, m):
      board[i][j] = 'x'
  return board

play_chomp()
```

This implementation works for boards in cases:
- $1 \times n$ or $m \times 1$: This is a trivial case because since we are the player to make the first move, we can just go to the right by 1 or up by 1 and this forces player 2 to eat the poison
- The $2 \times n$ or $m \times 2$ case uses the strategy of always having the board turn into a rectangle with 1 less piece on the upper right. If the strategy keeps going and going, we will eventually get to when player 2 has to eat the chocolate
- The $m = n$ case is unique in its strategy. The strategy goes that player 1 will condense the board down into an L-shaped board. After this, player 1 matches player 2's moves by whatever player 2 does, player 1 does the equivalent version of that move flipped across the 45-degree axis. This makes the board always symmetric at the start of player 2's turn, gauranteeing that an extra move after their turn is possible.
